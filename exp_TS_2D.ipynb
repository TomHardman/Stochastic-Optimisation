```python
fig, ax = plt.subplots(1, 3, figsize=(8, 3))

for i, data in enumerate(data_list_actual):
    ax[i].hist(data_list_actual[i], bins=10)
    ax[i].set_title(label_list[i])
    ax[i].vlines(0.215, ymin=0, ymax=30, color='magenta', label='0.215')
    ax[i].vlines(0.263, ymin=0, ymax=30, color='r', label='0.263')
    ax[i].vlines(0.365, ymin=0, ymax=30, color='orange', label='0.365')
    if i == 2:
        ax[i].legend(loc='upper left')

plt.show()
```
