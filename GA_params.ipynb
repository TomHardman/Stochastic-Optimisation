```python
# vary parents ratio
# notes low p ratio, very little movement in blue curve, and low end value -> premature convergence
# medium, trade off exploitation and exploration, finds a higher maximum
# high p ratio, more movement in blue curve and red curve, but due to high parents ratio, introduction of suboptimal genes, overall end solution found worse
model = None

np.random.seed(8)
varbound = np.array([[0, 10]]*8)

vartype = np.array(['real']*8)
p_list = [0.1, 0.5, 0.8]
for p in p_list:

    algo_param = {'max_num_iteration': 10000,\
                    'population_size':50,\
                    'mutation_probability':0.1,\
                    'elit_ratio': 0.01,\
                    'crossover_probability': 0.5,\
                    'parents_portion': p,\
                    'crossover_type':'uniform',\
                    'max_iteration_without_improv': None,
                    'scheme':'prop'}

    model = ga(dimension=8, variable_type='real', variable_boundaries=varbound, algorithm_parameters=algo_param)
    model.run()
    model=None
```

```python
# vary crossover probability
# low c, premature convergence, unable to search space effectively
# mediu
model = None

np.random.seed(8)
varbound = np.array([[0, 10]]*8)

vartype = np.array(['real']*8)
c_list = [0.001, 0.6, 1]
for c in c_list:

    algo_param = {'max_num_iteration': 10000,\
                    'population_size':50,\
                    'mutation_probability':0.1,\
                    'elit_ratio': 0.01,\
                    'crossover_probability': c,\
                    'parents_portion': 0.3,\
                    'crossover_type':'uniform',\
                    'max_iteration_without_improv': None,
                    'scheme':'prop'}

    model = ga(dimension=8, variable_type='real', variable_boundaries=varbound, algorithm_parameters=algo_param)
    model.run()
    model=None
```

```python
# vary mutation probability
# low c, pure cross over leaves areas of search space unreachable, therefore ecertain paths cannot be explored, these paths may lead to higher optimum. can see a lower final value.
#also average if very close to the best solution, therefore very close to elites, signifies low exploration
# mediuc, increased exploration, greater variation in average score -> higehr epxloration, improved final value
# high mutation, a lot of exploration and no premature convergence, however, destruction of good gene information due to constant mutation, information given by search not utilised
model = None

np.random.seed(8)
varbound = np.array([[0, 10]]*8)

vartype = np.array(['real']*8)
m_list = [0.001, 0.1, 0.8]
for m in m_list:

    algo_param = {'max_num_iteration': 10000,\
                    'population_size':50,\
                    'mutation_probability': m,\
                    'elit_ratio': 0.01,\
                    'crossover_probability': 0.5,\
                    'parents_portion': 0.6,\
                    'crossover_type':'uniform',\
                    'max_iteration_without_improv': None,
                    'scheme':'prop'}

    model = ga(dimension=8, variable_type='real', variable_boundaries=varbound, algorithm_parameters=algo_param)
    model.run()
    model=None
```
