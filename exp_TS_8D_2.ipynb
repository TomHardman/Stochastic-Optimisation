```python
from TS import run_multiple_ts, grid_search_alpha_subset, grid_search_alpha_pri_period
from util import keanes_bump_ts
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%load_ext autoreload
%autoreload 2
```

```python
algo_param  = {'search_method': 'exhaustive', 
                'N': 7,
                'M': 4, 
                'i_count': 10,
                'd_count': 15,
                'r_count': 25,
                'subset_size': 4,
                'prioritisation_period': 10,
                'initial_step_size': 3,
                'reduce_factor': 0.9,
                'n_cells': 2}
f = keanes_bump_ts
best_arr_ex = run_multiple_ts(50, f, 8, algo_param, log=False)
```

```python
algo_param  = {'search_method': 'random_subset', 
                'N': 7,
                'M': 4, 
                'i_count': 10,
                'd_count': 15,
                'r_count': 25,
                'subset_size': 3,
                'prioritisation_period': 10,
                'initial_step_size': 3,
                'reduce_factor': 0.94,
                'n_cells': 2}
f = keanes_bump_ts
best_arr_rss = run_multiple_ts(50, f, 8, algo_param, log=False)
```

```python
algo_param  = {'search_method': 'successive_random_subset', 
                'N': 7,
                'M': 4, 
                'i_count': 10,
                'd_count': 15,
                'r_count': 25,
                'subset_size': 4,
                'prioritisation_period': 10,
                'initial_step_size': 3,
                'reduce_factor': 0.9,
                'n_cells': 2}
f = keanes_bump_ts
best_arr_srss = run_multiple_ts(50, f, 8, algo_param, log=False)
```

```python
fig, ax = plt.subplots()
data_list = [best_arr_ex, best_arr_rss, best_arr_srss]
data_list_actual = []
for arr in data_list:
    data_list_actual.append([sol[1] for sol in arr])

label_list = ['exhaustive', 'random subset', 'successive random subset']
ax.boxplot(data_list_actual, vert=False, patch_artist=True, labels=label_list, whis=30)
plt.show()
```
